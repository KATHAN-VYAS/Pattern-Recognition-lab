Boundry detection using contour tracing

contour: curve joining all continuous points along the boundary of an object that have the same intensity (color). For binary images, this means tracing the white shapes on a black background (or vice versa).

```python
'''
contour detection vs edge detection:https://how.dev/answers/edge-detection-vs-contour-detection-in-computer-vision

We can define an edge to be a change in pixel or image intensity.


steps: 

1. read the image and convert to binary
2. use cv2.findContours()
3. draw contour on copy of image
4. display
'''
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread(r'D:\kathan\python_codes\7th sem\Pattern-Recognition-lab\image copy.png', cv2.IMREAD_GRAYSCALE)
print("image read successfully")

```

```python
#converting to binary img:
_, bin = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(bin)
```

```python
# Detect contours
contours, hierarchy = cv2.findContours(bin, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# draw contour on copy of image:

img_contours = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
cv2.drawContours(img_contours, contours, -1, (0, 255, 0), 2)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(bin, cmap='gray')
plt.title("Binary Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_contours)
plt.title("Detected Contours")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
'''
Types of contouring:

1. cv2.RETR_EXTERNAL -  ignores nested (internal) contours entirely.
                        Useful for blob detection, counting distinct objects, or applications where you just need a mask of external shapes.
2. cv2.RETR_LIST - Finds all contours in the image (outer + inner) but doesnt build a parent-child 
                    hierarchy between them. when dont care about their relationships (e.g., which is inside which).
3. cv2.RETR_TREE - When you have nested structures, and knowing which contours are inside or outside others is essential.
4. cv2.RETR_CCOMP - Finds all contours and organizes them into two levels:
            
            Level 1 - outermost contours.
            Level 2 - holes inside each outer contour.
            
            When you need to differentiate between filled objects and holes inside them, but dont need multi-level nesting.
'''

```
